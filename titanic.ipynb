```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic
```

```python
titanic.head()
```

```python
titanic.duplicated()
```

```python
titanic.isnull().sum()
```

```python
titanic.describe()
```

```python
titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode(),inplace=True)
titanic
```

```python
titanic.drop(columns=['Cabin'],inplace=True)
```

```python
titanic.info()
```

```python
titanic.dropna(inplace=True)
```

```python
titanic['Sex'] = titanic['Sex'].map({"male":0,"female":1})
titanic['Embarked'] = titanic['Embarked'].map({"S":1,"C":2,"Q":3})
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic.columns
```

```python
X = titanic[[ 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
y = titanic['Survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred[:10]
```

```python
y_test[:10]
```

```python
pred_df = pd.DataFrame({
    "Actual":y_test,
    "Predicted":y_pred
})
pred_df
```

```python
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

```python
sns.heatmap(conf_mat,annot=True,fmt='d')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
print("Accuracy: ",accuracy_score(y_test,y_pred)*100)
print("Precision: ",precision_score(y_test,y_pred)*100)
print("Recall: ",recall_score(y_test,y_pred)*100)
print("Classfication Report:\n ",classification_report(y_test,y_pred))
```
